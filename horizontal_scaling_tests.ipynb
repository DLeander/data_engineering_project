```python
# Test 1.
# 1 core on one worker node.

from pyspark.sql import SparkSession
import time

start = time.time()

# Create the spark session with the correct maximum cores
# and number of executor cores that corresponds to the test.
spark_session = SparkSession\
        .builder\
        .master("spark://192.168.2.133:7077") \
        .appName("horizontal_scaling_test_1")\
        .config("spark.dynamicAllocation.enabled", False)\
        .config("spark.dynamicAllocation.shuffleTracking.enabled",False)\
        .config("spark.shuffle.service.enabled", False)\
        .config("spark.cores.max", 1)\
        .config("spark.executor.cores",1)\
        .getOrCreate()
spark_context = spark_session.sparkContext

# Read the pre-processed data (csv file)
df = spark_session.read\
    .option("header", "true")\
    .csv('hdfs://192.168.2.133:9000/SongCSV.csv')

# Count how many Songs there are for each Album.
df.groupBy("AlbumName").count().orderBy('count', ascending=False).show()

# Release cores and stop application.
spark_context.stop()

end = time.time()

runtime_test_1 = end-start
```
